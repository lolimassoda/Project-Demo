# Linear Regression Models – mtcars Dataset (FINAL SUBMISSION)

This notebook creates 4 linear regression models using the mtcars dataset.

Models:
1. mpg ~ hp  
2. mpg ~ wt  
3. mpg ~ gear  
4. mpg ~ hp + wt + gear + am  

Methods used:
- sklearn LinearRegression  
- Normal Equation  
- Gradient Descent  

Model fit evaluated using R² and F-statistic.


```python

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

df = pd.read_csv("mtcars (1).csv")
df.columns = df.columns.str.lower()
df.head()
```

```python

y = df['mpg']

x1 = df[['hp']]
x2 = df[['wt']]
x3 = df[['gear']]
x4 = df[['hp','wt','gear','am']]

models = {
"Model 1 (HP)": x1,
"Model 2 (WT)": x2,
"Model 3 (Gear)": x3,
"Model 4 (HP, WT, Gear, AM)": x4
}
```

```python

def sklearn_lm(x, y):
    model = LinearRegression()
    model.fit(x, y)
    return model.intercept_, model.coef_

def normal_eq_lm(x, y):
    X = np.c_[np.ones(len(x)), x]
    beta = np.linalg.pinv(X.T @ X) @ X.T @ y
    return beta

def gd_lm(x, y, lr=0.001, epochs=10000):
    X = np.c_[np.ones(len(x)), x]
    beta = np.zeros(X.shape[1])
    for _ in range(epochs):
        pred = X @ beta
        error = pred - y
        beta -= lr * (X.T @ error) / len(y)
    return beta
```

```python

for name, x in models.items():
    print("\n======================")
    print(name)
    print("======================")
    
    print("Sklearn:", sklearn_lm(x,y))
    print("Normal Equation:", normal_eq_lm(x,y))
    print("Gradient Descent:", gd_lm(x,y))

    X_sm = sm.add_constant(x)
    model = sm.OLS(y, X_sm).fit()

    print("R-squared:", model.rsquared)
    print("F-statistic:", model.fvalue)
    print("p-value:", model.f_pvalue)
```


## Interpretation

Model 1 (hp):
Horsepower has a negative relationship with mpg. Higher horsepower usually lowers fuel efficiency.

Model 2 (wt):
Weight strongly affects mpg. Heavier vehicles generally consume more fuel.

Model 3 (gear):
Gear alone is a weak predictor of mpg.

Model 4 (hp, wt, gear, am):
This model fits best because multiple variables improve prediction accuracy.
