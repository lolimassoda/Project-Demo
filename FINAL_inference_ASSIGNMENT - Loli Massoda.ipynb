# Inference Notebook

Use the saved model to predict heart disease for a new patient.


```python

import numpy as np
import pandas as pd
import joblib
```

```python

# Load saved model pipeline
model_bundle = joblib.load("best_heart_model.pkl")

preprocessor = model_bundle["preprocessor"]
selector = model_bundle["selector"]
model = model_bundle["model"]

print("Model and preprocessing objects loaded.")
```

```python

# Define a new unseen sample (example values)
# IMPORTANT: This should NOT be a row copied from the training/test set.
new_sample = pd.DataFrame([{
    "age": 58,
    "sex": 1,
    "cp": 2,
    "trestbps": 140,
    "chol": 240,
    "fbs": 0,
    "restecg": 1,
    "thalach": 150,
    "exang": 0,
    "oldpeak": 1.2,
    "slope": 1,
    "ca": 0,
    "thal": 2
}])

new_sample
```

```python

# Apply preprocessing and feature selection (same as training pipeline)
X_new_processed = preprocessor.transform(new_sample)
X_new_selected = selector.transform(X_new_processed)

X_new_selected.shape
```

```python

# Run prediction
prob = model.predict_proba(X_new_selected)[0, 1]
pred = model.predict(X_new_selected)[0]

print(f"Predicted probability of heart disease: {prob:.2f}")
print("Predicted class (1 = heart disease, 0 = no heart disease):", int(pred))
```
